```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Poisson's equation

ncells = np.array([1000, 4000, 16000, 64000, 256000, 1024000], dtype=float)
times = np.array([0.45, 0.52, 1.49, 9.09, 48.36, 322.97])

plt.loglog(ncells, times, 'o--')
plt.loglog(ncells, times[0]/ncells[0] * ncells, 'C2:', label='Linear')
plt.loglog(ncells, times[0]/ncells[0]**2 * ncells**2, 'C3:', label='Quadratic')
plt.xlabel('Number of cells')
plt.ylabel('Computation time (s)')
plt.title("Poisson's Equation")
plt.legend(loc=0)
plt.savefig('poisson-scale.png')
plt.show()
```

```python
# Heat equation

ncells = np.array([4000, 8000, 16000, 32000, 64000, 128000, 256000], dtype=float)
times = np.array([0.40, 1.71, 6.79, 24.50, 107.69, 215.84, 509.70])

plt.loglog(ncells, times, 'o--')
plt.loglog(ncells, times[0]/ncells[0] * ncells, 'C2:', label='Linear')
plt.loglog(ncells, times[0]/ncells[0]**2 * ncells**2, 'C3:', label='Quadratic')
plt.xlabel('Number of cells')
plt.ylabel('Computation time (s)')
plt.title('Heat Equation')
plt.legend()
plt.savefig('heat-scale.png')
plt.show()
```
